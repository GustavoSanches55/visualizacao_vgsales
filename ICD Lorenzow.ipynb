```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv("vgsales.csv")
df.dropna(how = "any",inplace = True)
print (df)
```

```python
#Dados
consoles_year = df.pivot_table('Global_Sales', index='Year', columns='Platform', aggfunc=sum)
main_consoles = consoles_year[['X360', 'PS', 'Wii', 'WiiU', 'PS2', 'PS3']]
#Gráfico
main_consoles.plot(subplots=True, figsize=(16, 16), grid=False, title="Global sales per year")
```

Observando o gráfico acima vemos que os consoles como PS2, PS3, e Xbox360 permanecem ativos por mais tempo no mercado que o Wii ou WiiU. Esses consoles apresentam um maior número de vendas de jogos deles.

```python
titles_global = df.pivot_table('Global_Sales', index='Year', columns='Name', aggfunc=sum)
titles_global
```

# Como é a relação entre Gêneros dos Jogos e as regiões de vendas?

```python
var = df.groupby(['Genre']).sum()
var
```

É possível perceber com essa tabela que existe uma enorme diferença na popularidade dos genêros da industria entre as diversas regiões do globo.  

```python
#Organizando Dados
df2 = df[df.columns.difference(['Rank', 'Year', 'Global_Sales'])]

var = df2.groupby(['Genre']).sum().sort_values('NA_Sales', ascending=True)
#Plotando Gráfico
var.plot(kind='barh', 
                   stacked=True, 
                   color=['salmon','royalblue','gold','forestgreen','pink','crimson','cyan','slategray','midnightblue','darkred','burlywood','darkolivegreen'], 
                   grid=False, 
                   figsize=(15,8),title="Vendas de genêros por região")
```

Esse gráfico põe em evidência a diferença da popularidade dos genêros entre as diferentes regiões analísadas. Se observarmos com atenção fica claro uma prefêrencia por jogos de Ação e Esporte nos Estados Unidos e Europa, e por jogos de Role_playing no Japão. Entretanto como evoluiram as vendas através dos anos?

```python
#Organizando Dados
titles_JP = df.pivot_table('JP_Sales', index='Year', columns='Genre', aggfunc=sum)
#Plotando gráfico
fig, ax = plt.subplots(figsize=(15,8))
sns.heatmap(titles_JP.transpose())
```

```python
#Organizando Dados
titles_NA = df.pivot_table('NA_Sales', index='Year', columns='Genre', aggfunc=sum)
#Plotando gráfico
fig, ax = plt.subplots(figsize=(15,8))
sns.heatmap(titles_NA.transpose())
```

Se observarmos ambos os gráficos vemos que tanto na América do Norte quanto no Japão, os genêros que faziam sucesso no começo não são os mesmos que faem sucesso agora. No Japão vemos por exemplo uma presença muito mais forte de jogos de estilo Plataforma na década de 1980 do que de RPGs. 

# Como se sobresaem títulos de franquias grandes?

```python
#Organizando dados
Gta_titles = df.pivot_table('Global_Sales', index='Year', columns='Name', aggfunc=sum)
Gta_titles
GTA_console = Gta_titles[['Grand Theft Auto V', 'Grand Theft Auto: Chinatown Wars', 'Grand Theft Auto IV', 'Grand Theft Auto: San Andreas', 'Grand Theft Auto: Vice City', 'Grand Theft Auto: Liberty City Stories', 'Grand Theft Auto III']]
GTA_console
data = {'Grand Theft Auto V': 55.92,
        'Grand Theft Auto: Chinatown Wars': 2.43,
        'Grand Theft Auto IV': 22.47,
        'Grand Theft Auto: San Andreas': 23.86,
        'Grand Theft Auto: Liberty City Stories': 11.26,
        'Grand Theft Auto: Vice City': 16.19,
        'Grand Theft Auto III': 13.11,}
group_data = list(data.values())
group_names = list(data.keys())
group_mean = np.mean(group_data)
#Plotando o gráfico
fig, ax = plt.subplots()
ax.barh(group_names, group_data)
labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')
ax.set(xlim=[0, 60], xlabel='Global Sales', ylabel='Jogos ordenados por lançamento',
       title='Franquia GTA')
```

Observando o gráfico acima vemos que a franquia de ação Grand theft auto, cresce progressivamente a cada lançamento com excessão dos título Liberty City Stories, ChinaTown Wars - jogos spin-off para consoles portáteis -, e Grand Theft Auto IV, que não é maior que o predecessor por uma pequena margem.
Entretanto isso é válido para jogos de todos os genêros?

```python
poki = df[df.Name.str.startswith('Pokemon')].sort_values('Year', ascending=True)
poki_top5 = poki[['Name', 'Global_Sales']].head(15).set_index('Name')
poki_top5.plot(kind='bar', figsize=(8,8), grid=False, color='royalblue', legend=False).get_xticklabels()
plt.title('Vendas globais da franquia Pokemon') 
plt.xlabel('Títulos') 
plt.ylabel('Vendas Globais')
plt.show
```

Observando o gráfico a seguir vemos que a saga Pokemon se sobresai de forma oposta a Grand Thef Auto. Se focarmos somente nos lançamentos principais da saga, Pokemon Red/Blue, Gold/Silver, e Ruby/Saphire, se torna evidente que a cada lançamento a franquia vende menos. Essa evolução se mantém entre todas as regiões do globo?

```python
poki = df[df.Name.str.startswith('Pokemon')].sort_values('Year', ascending=True)
poki_top5 = poki[['Name', 'JP_Sales']].head(15).set_index('Name')
poki_top5.plot(kind='bar', figsize=(8,8), grid=False, color='crimson', legend=False).get_xticklabels()
plt.title('Venda dos títulos da saga Pokemon no Japão') 
plt.xlabel('Títulos') 
plt.ylabel('Vendas no Japão')
plt.show
```

Se observarmos o gráfico acima podemos ver que no próprio páis onde o jogo foi produzido a relação de de vendas entre os jogos não muda tanto. Uma das muitas diferenças entre essas duas franquias são os consoles nos quais elas estão presentes. O sucesso de uma franquia pode ser dependente do console ao qual ela pertence?
