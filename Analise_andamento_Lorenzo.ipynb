```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
import seaborn as sns

%reload_ext autoreload
%autoreload 2
```

```python
df = pd.read_csv (r'C:\Users\Pichau\Desktop\vgsales.csv')
df.dropna(how = "any",inplace = True)
print (df)
```

```python
sns.distplot(df['Global_Sales'])
```

```python
sns.boxplot(x='Platform', y='Global_Sales', data= df)
```

```python
df[df.Platform.str.startswith('X', 'P', 'W')]
```

```python
main_consoles = df[df.Platform== 'X360']
main_consoles
```

```python
consoles_year = df.pivot_table('Global_Sales', index='Year', columns='Platform', aggfunc=sum)
consoles_year
```

```python
main_consoles = consoles_year[['X360', 'XOne', 'Wii', 'WiiU', 'PS2', 'PS3']]
main_consoles.plot(subplots=True, figsize=(12, 12), grid=False, title="Global sales per year")
```

```python
titles_global = df.pivot_table('Global_Sales', index='Platform', columns='Name', aggfunc=sum)
titles_global
```

```python
GTA_console = titles_global[['Grand Theft Auto V', 'Grand Theft Auto IV', 'Grand Theft Auto: San Andreas', 'Grand Theft Auto: Vice City', 'Grand Theft Auto: Liberty City Stories', 'Grand Theft Auto III']]
GTA_console.plot(subplots=True, figsize=(12, 12), grid=False, title="Grand Theft auto Sales per console")
```

```python
var = df.groupby(['Genre']).sum()
var
```

```python
df2 = df[df.columns.difference(['Rank', 'Year', 'Global_Sales'])]

var = df2.groupby(['Genre']).sum()
var_t = var.T
var_t
```

```python
var_t.plot(kind='bar', 
                   stacked=True, 
                   color=['salmon','royalblue','gold','forestgreen','pink','crimson','cyan','slategray','midnightblue','darkred','burlywood','darkolivegreen'], 
                   grid=False, 
                   figsize=(15,8),title="Genre Sales per location")
```
